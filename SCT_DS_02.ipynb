```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
gender_submission = pd.read_csv('gender_submission.csv')
```

```python
# Display basic information
print("Train Data Info:")
print(train.info())

print("\nTrain Data Description:")
print(train.describe())

print("\nTrain Data Value Counts (Survived):")
print(train['Survived'].value_counts())
```

```python
# Checking missing values
print("\nMissing Values:")
print(train.isnull().sum())
```

```python
# Visualizations
sns.set(style="whitegrid")
```

```python
# 1. Histograms
train.hist(figsize=(12,10))
plt.suptitle("Histograms of Train Data", fontsize=16)
plt.show()
```

```python
# 2. Boxplots
plt.figure(figsize=(8,6))
sns.boxplot(x='Survived', y='Age', data=train)
plt.title('Boxplot of Age by Survival')
plt.show()
```

```python
# 3. Scatterplots
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train)
plt.title('Scatterplot of Age vs Fare Colored by Survival')
plt.show()
```

```python
# 4. Pairplot
sns.pairplot(train[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']], hue='Survived')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Only select numerical columns for plotting
numeric_cols = train.select_dtypes(include=[np.number])

# Now plot histograms safely
numeric_cols.hist(figsize=(12,10))
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()

```
